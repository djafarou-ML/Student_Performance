```python
df.drop(['student_id', 'grade'], axis='columns', inplace=True)
df
```

```python
df = df.select_dtypes(include='number')
df
```

```python
plt.Figure(figsize=(15,8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, square=True)
plt.title("Correlation")
plt.show()
```

```python
plt.figure(figsize=(5, 6)) 
plt.scatter(df['weekly_self_study_hours'], df['total_score'])
plt.xlabel("weekly_self_study_hours")
plt.ylabel("total_score")
plt.title("Corrélation entre le nombre d'heures d'études et le score")
plt.show()
```

```python
plt.figure(figsize=(5, 6))  
plt.scatter(df['attendance_percentage'], df['total_score'])
plt.xlabel("attendance_percentage")
plt.ylabel("total_score")
plt.title("Corrélation entre la presence et le score")
plt.show()
```

```python
plt.figure(figsize=(5, 6)) 
plt.scatter(df['class_participation'], df['total_score'])
plt.xlabel("class_participation")
plt.ylabel("total_score")
plt.title("Corrélation entre la participation et le score")
plt.show()
```
