```python
df.drop(['student_id', 'grade'], axis='columns', inplace=True)
df
```

**ANALYSE DE CORRÉLATION**

```python
plt.Figure(figsize=(15,8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.5f', square=True)
plt.title("Correlation")
plt.show()
```

```python
plt.figure(figsize=(5, 6)) 
plt.scatter(df['weekly_self_study_hours'], df['total_score'])
plt.xlabel("weekly_self_study_hours")
plt.ylabel("total_score")
plt.title("Corrélation entre le nombre d'heures d'études et le score")
plt.show()
```

```python
plt.figure(figsize=(5, 6)) 
plt.scatter(df['class_participation'], df['total_score'])
plt.xlabel("class_participation")
plt.ylabel("total_score")
plt.title("Corrélation entre la participation et le score")
plt.show()
```

```python
sns.boxplot(x=df['total_score'])
plt.show()
```

**PRÉPARATION DU MODÈLE**

```python
X = df.drop(['total_score', 'class_participation', 'attendance_percentage'], axis=1)
X
```

```python
Y = df['total_score']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

**ÉVALUATION ET PRÉDICTIONS**

```python
test_predict = model.predict(X_test)
test_predict
```

```python
X_test
```

```python
Y_test
```

```python
df['total_score'].mean()
```

```python
sns.histplot(data=df, x='total_score', bins=20)
plt.show()
```

```python
mean_absolute_error(Y_test, test_predict)
```

```python
plt.scatter(Y_test, test_predict)
plt.xlabel("Valeurs réelles")
plt.ylabel("Prédictions")
plt.show()
```

**VISUALISATION DES RÉSULTATS FINAUX**

```python
plt.figure(figsize=(10, 6))
# On affiche les points réels du jeu de test
plt.scatter(X_test, Y_test, color='blue', alpha=0.5, label='Données réelles')

# On affiche la droite de prédiction
plt.plot(X_test, test_predict, color='red', linewidth=3, label='Ligne de régression')

plt.title('Régression Simple : Heures d\'étude vs Score')
plt.xlabel('Heures d\'étude hebdomadaires')
plt.ylabel('Score Total')
plt.legend()
plt.show()
```

```python
test_residual = Y_test - test_predict
test_residual
```

```python
test_residual.describe()
```

```python
outliers = test_residual[abs(test_residual) > 30]
outliers
```
